```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Read energy file:

```python
step, time, KineticE, Temp, PotentialE, ConsQty, usedtime = np.loadtxt("NPT-1.ener", unpack=True)
```

```python
# Plot Kinetic Energy
plt.plot(time, KineticE)
plt.title("Kinetic Energy")
plt.ylabel("Energy / hartrees")
plt.xlabel("Simulation time / fs")
plt.savefig("KineticE.png")
```

```python
plt.plot(time, PotentialE)
plt.title("Potential Energy")
plt.ylabel("Energy / hartrees")
plt.xlabel("Simulation time / fs")
plt.savefig("PotentialE.png")
```

```python
plt.plot(time, Temp)
plt.title("Temperature")
plt.ylabel("Temperature / K")
plt.xlabel("Simulation time / fs")
plt.savefig("Temp.png")
```

### Read energy file:

```python
step, time, AX, AY, AZ, BX, BY, BZ, CX, CY, CZ, Volume = np.loadtxt("NPT-1.cell", unpack=True)
```

```python
plt.plot(time, Volume)
plt.title("Simulation box volume")
plt.ylabel("Volume / Angstrom^3")
plt.xlabel("Simulation time / fs")
plt.savefig("Volume.png")
```

```python
plt.plot(time, AX, label="X-axis")
plt.plot(time, BY, label="Y-axis")
plt.plot(time, CZ, label="Z-axis")
plt.title("Box size")
plt.ylabel("Distance / Angstrom")
plt.xlabel("Simulation time / fs")
plt.savefig("BoxSize.png")
```
